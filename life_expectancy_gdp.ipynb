```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.api as sm
from scipy.stats import pearsonr

df = pd.read_csv('all_data.csv')
print(df.head(10))

print(df.Country.unique())
print(df.Year.unique())
```

```python
df = df.rename({"Life expectancy at birth (years)":"Life_Expectancy"}, axis = "columns")
df.head()
```

```python
chile_data = df[df['Country'] == 'Chile']
print(chile_data.head())

sns.lineplot(x = chile_data.Year, y = chile_data.Life_Expectancy)
plt.title('Life Expectancy in Chile')
plt.show()
plt.clf()

sns.lineplot(x = chile_data.Year, y = chile_data.GDP)
plt.title('GDP in Chile')
plt.show()
plt.clf()

model = sm.OLS.from_formula('Life_Expectancy ~ GDP', data = chile_data)
results = model.fit()
print(results.params)

plt.scatter(chile_data.GDP, chile_data.Life_Expectancy)
plt.plot(chile_data.GDP, results.params[1]*chile_data.GDP + results.params[0])
plt.title('GDP vs Life Expectancy in Chile')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.show()
plt.clf()

corrcoeff_chile = np.corrcoef(chile_data.GDP, chile_data.Life_Expectancy)[0, 1]
print("Correlation coefficient of GDP and life expectancy in Chile: " + str(corrcoeff_chile))

```

* There is a strong positive correlation between GDP and life expectancy in Chile.
* There is a decline in GDP and life expectancy between 2008 and 2010.
* There is also a decline in GDP from 2013 and above.

```python
# Calculate average life expectancy for each country and plot it on the bar
chile_life = int(np.sum(chile_data.Life_Expectancy))/len(chile_data.Year)
china_life = int(np.sum(china_data.Life_Expectancy))/len(china_data.Year)
germany_life = int(np.sum(germany_data.Life_Expectancy))/len(germany_data.Year)
usa_life = int(np.sum(usa_data.Life_Expectancy))/len(usa_data.Year)
mexico_life = int(np.sum(mexico_data.Life_Expectancy))/len(mexico_data.Year)
zimbabwe_life = int(np.sum(zimbabwe_data.Life_Expectancy))/len(zimbabwe_data.Year)

country = ['Chile', 'China', 'Germany', 'USA', 'Mexico', 'Zimbabwe']
life = [chile_life, china_life, germany_life, usa_life, mexico_life, zimbabwe_life]

plt.bar(country, life)
plt.title("Average Life Expectancy in Each Country")
plt.xlabel("Country")
plt.ylabel("Life Expectancy")
plt.show()
plt.clf()
```

* It is observed that all countries have similar life expectancy except Zimbabwe.

```python
# Calculate average GDP for each country and plot it in a bar
chile_gdp = int(np.sum(chile_data.GDP))/len(chile_data.Year)
china_gdp = int(np.sum(china_data.GDP))/len(china_data.Year)
germany_gdp = int(np.sum(germany_data.GDP))/len(germany_data.Year)
usa_gdp = int(np.sum(usa_data.GDP))/len(usa_data.Year)
mexico_gdp = int(np.sum(mexico_data.GDP))/len(mexico_data.Year)
zimbabwe_gdp = int(np.sum(zimbabwe_data.GDP))/len(zimbabwe_data.Year)

gdp = [chile_gdp, china_gdp, germany_gdp, usa_gdp, mexico_gdp, zimbabwe_gdp]

plt.bar(country, gdp)
plt.title("Average GDP in Each Country")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.show()
plt.clf()
```

* USA has the highest GDP meanwhile Chile and Zimbabwe have the lowest GDP
